```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
initial_theta=0
final_theta=75
t_total=3
alpha=10
A=np.zeros(6)
t=np.arange(0,3,0.01)
a,b,c,d,e,f = sym.symbols('a,b,c,d,e,f')
eq1 = sym.Eq(a,0)
eq2 = sym.Eq(b,0)
eq3 = sym.Eq(2*c,10)
eq4 = sym.Eq(27*d+81*e+243*f,30)
eq5 = sym.Eq(27*d+108*e+405*f,-30)
eq6 = sym.Eq(18*d+108*e+540*f,-20)
result = sym.solve([eq1,eq2,eq3,eq4,eq5,eq6],(a,b,c,d,e,f))
result=list(result.values())
for i in range(len(result)):
  A[i]=result[i]
theta=A[0]+A[1]*t+A[2]*t*t+A[3]*t*t*t+A[4]*t*t*t*t+A[5]*t*t*t*t*t
w=A[1]+2*A[2]*t+3*A[3]*t*t+4*A[4]*t*t*t+5*A[5]*t*t*t*t
acc=2*A[2]+6*A[3]*t+12*A[4]*t*t+20*A[5]*t*t*t

print("The Coefficients of 5th order polynomial equations are: ")
print("a0= ",result[0])
print("a1= ",result[1])
print("a2= ",result[2])
print("a3= ",result[3])
print("a4= ",result[4])
print("a5= ",result[5])

plt.plot(t,theta)
plt.xlabel("Time")
plt.ylabel("Theta")
plt.title("Plot between Time and Theta")

plt.plot(t,w)
plt.xlabel("Time")
plt.ylabel("Angular Velocity")
plt.title("Plot between Time and Angular Velocity")

plt.plot(t,acc)
plt.xlabel("Time")
plt.ylabel("Angular Acceleration")
plt.title("Plot between Time and Angular Acceleration")
```
